**Importing**

```python
import pandas as pd
import numpy as np

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report,confusion_matrix
```

**Importing Dataset**

```python
emails = pd.read_csv('/content/drive/MyDrive/Datasets/Email spam or Ham/emails.csv')
print(emails.head())
print(emails['Prediction'].unique())
```

```python
emails.shape
```

```python
emails.describe(include = 'all')
```

```python
x = emails.drop(['Prediction','Email No.'], axis = 1)
y = emails['Prediction']
print(x.head(2))
print(y.head(5))
```

**Splitting Dataset**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
print(x_train.shape)
print(x_test.shape)
```

**Training the data**

```python
mlp = MLPClassifier(hidden_layer_sizes=(3,2), activation = 'logistic')
```

```python
mlp.fit(x_train,y_train)
```

**Testing**

```python
pred = mlp.predict(x_test)
pred
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
